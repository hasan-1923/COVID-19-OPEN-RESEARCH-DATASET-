```python
f, axs = plt.subplots(3,2, figsize = (14,14), constrained_layout = True)

sort_de = data.sort_values(by = ['Deaths / 100 Recovered'], ascending = False)
sort_de = sort_de.head(10)
sort_co = data.sort_values(by = ["Confirmed"], ascending = False)
sort_co = sort_co.head(10)
sort_dea = data.sort_values(by = ['Deaths'], ascending = False)
sort_dea = sort_dea.head(10)
sort_rec = data.sort_values(by = ['Recovered'], ascending = False)
sort_rec = sort_rec.head(10)
sort_ac = data.sort_values(by = ['Active'], ascending = False)
sort_ac = sort_ac.head(10)
sort_new = data.sort_values(by = ['New cases'], ascending = False)
sort_new = sort_new.head(10)

ax=sns.barplot(x="Country/Region", y="Deaths / 100 Recovered", data=sort_de, ax = axs[0, 0])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="Country/Region",y="Confirmed", data=sort_co, ax = axs[0, 1])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="Country/Region",y="Deaths", data=sort_de, ax = axs[1, 0])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="Country/Region",y="Recovered", data=sort_rec, ax = axs[1, 1])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="Country/Region",y="Active", data = sort_ac, ax = axs[2, 0])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="Country/Region",y="New cases", data=sort_new, ax = axs[2, 1])
ax.tick_params(axis = 'x', rotation = 45)
```
